```python
# Spaghetti code and hashed comment is making loops with 'for' iteration

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Year'] = df['InvoiceDate'].dt.year.astype('int32')
df['Month'] = df['InvoiceDate'].dt.month_name()
df['Day'] = df['InvoiceDate'].dt.day.astype('int32')
df['Hour'] = df['InvoiceDate'].dt.hour.astype('int32')
df['Minute'] = df['InvoiceDate'].dt.minute.astype('int32')
df['Second'] = df['InvoiceDate'].dt.second.astype('int32')
df['Weekday'] = df['InvoiceDate'].dt.strftime('%A')
ymdhmsw = ['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'Weekday']
# for col in ymdhmsw:
#     df[col] = df['InvoiceDate'].dt.__getattribute__(col.lower())
#     df.insert(5, col, df.pop(col))
df.insert(loc=5, column='Weekday', value=df.pop('Weekday'))
df.insert(loc=5, column='Second', value=df.pop('Second'))
df.insert(loc=5, column='Minute', value=df.pop('Minute'))
df.insert(loc=5, column='Hour', value=df.pop('Hour'))
df.insert(loc=5, column='Day', value=df.pop('Day'))
df.insert(loc=5, column='Month', value=df.pop('Month'))
df.insert(loc=5, column='Year', value=df.pop('Year'))
df
```

```python
#Regarding calculating recency (advise from prof. Kwon):
#creating timestamp from invoice date to next purchasing date

import datetime as dt
from datetime import datetime, timedelta
customer_retention = df.groupby('Customer ID')['InvoiceDate'].max().reset_index() # - The last transaction data will have 1 for recency.
customer_retention['CustomerRetention'] = customer_retention['InvoiceDate'] + dt.timedelta(days=1) # - last_day = the last transaction day + use 'timedelta' function to add one more day.
customer_retention['Recency'] = (customer_retention['CustomerRetention'] - df['InvoiceDate']).dt.days # - Then, recency = (last_day - InvoiceDate).dt.days, which will give you the data you want go get. The recency for the last transaction date is 1.
print(customer_retention)
```

```python
df = df.loc[(df['Quantity'] > 0) & (df['Price'] != 0)] #negative 'Q' (Return goods) and zero 'P' (Free gift) has been cleared.
df
```

```python
print("Number of transactions: ", df['Invoice'].nunique())
print("Number of products: ",df['StockCode'].nunique())
print("Number of customers:", df['Customer ID'].nunique())
groupby_invoice = pd.DataFrame(df.groupby('Invoice')['StockCode'].nunique()) #build the composition of baskets, group by 'Invoice' and classification for transaction, products, customers variable.
groupby_invoice['CustomersNo'] = df.groupby('Invoice')['Customer ID'].nunique()
groupby_invoice
```

```python
freq = df['InvoiceDate'].max()
r_df = df.groupby('Customer ID').agg({'InvoiceDate': lambda x: (freq - x.max()).days}) #recency measurement
r_df.rename(columns={'InvoiceDate': 'Recency'}, inplace=True)
f_df = df.groupby('Customer ID').agg({'Invoice': 'nunique'}) #frequency measurement
f_df.rename(columns={'Invoice': 'Frequency'}, inplace=True)
m_df = df.groupby('Customer ID').agg({'Price': 'sum'}) #monetary value measurement
m_df.rename(columns={'Price': 'MonetaryValue'}, inplace=True)
```

```python
r_f = pd.merge(r_df, f_df, on='Customer ID')
r_f_m = pd.merge(r_f, m_df, on='Customer ID')
r_f_m
```

```python
r_f_m.describe().T
```

```python
r_labels = range(1, 4) #take professsor's advice -- all labels should make array in 1, 2, 3. not 3, 2, 1.
f_labels = range(1, 4)
m_labels = range(1, 4)

r_groups = pd.qcut(r_df['Recency'], q=3, labels=r_labels) #quantative cutting into 3 groups (*slicing for 3:head, 2:body, 1:tail)
f_groups = pd.qcut(f_df['Frequency'], q=3, labels=f_labels)
m_groups = pd.qcut(m_df['MonetaryValue'], q=3, labels=m_labels)

rfm_df = pd.concat([r_groups, f_groups, m_groups], axis=1)
rfm_df.columns = ['R', 'F', 'M']
rfm_df['RFM_Group'] = rfm_df.apply(lambda x: str(x['R']) + str(x['F']) + str(x['M']), axis=1)

rfm_df
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
preprocessor = Pipeline([
  ("scaler", MinMaxScaler())
])
clusterer = Pipeline(
   [
       (
          "kmeans",
           KMeans(
               n_clusters= 3,
               init="k-means++",
               random_state=42,
           ),
       ),
   ]
)
pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)
pipe.fit(r_f_m[['Recency']])
preprocessed_data = pipe["preprocessor"].transform(r_f_m[['Recency']])
labels_r = pipe["clusterer"]["kmeans"].labels_
r_f_m["RecencyCluster"] = labels_r
r_f_m.head()
```

```python
pipe.fit(r_f_m[['Frequency']])
preprocessed_data = pipe["preprocessor"].transform(r_f_m[['Frequency']])
labels_f = pipe["clusterer"]["kmeans"].labels_
r_f_m["FrequencyCluster"] = labels_f
r_f_m.head()
```

```python
pipe.fit(r_f_m[['MonetaryValue']])
preprocessed_data = pipe["preprocessor"].transform(r_f_m[['MonetaryValue']])
labels_m = pipe["clusterer"]["kmeans"].labels_
r_f_m["MonetaryValueCluster"] = labels_m
r_f_m.head()
```

```python
r_f_m['RFMScore'] = r_f_m['RecencyCluster'] + r_f_m['FrequencyCluster'] + r_f_m['MonetaryValueCluster']
r_f_m['RFMScore'].unique()
```

```python
core_cust = r_f_m[r_f_m['RFMScore'] == 6].sort_values('Customer ID', ascending=True) #Core - Your Best Customers: Highly engaged customers who have bought the most recent, the most often, and generated the most revenue. (2+2+2)
loyal_cust = r_f_m[r_f_m['FrequencyCluster'] == 2].sort_values('Frequency', ascending=False) #Loyal - Your Most Loyal Customers: Customers who buy the most often from your store. (x+2+x)
whales_cust = r_f_m[r_f_m['MonetaryValueCluster'] == 2].sort_values('MonetaryValue', ascending=False) #Whales - Your Highest Paying Customers: Customers who have generated the most revenue for your store. (x+x+2)
newbie_cust = r_f_m[(r_f_m['RecencyCluster'] == 2) & (r_f_m['FrequencyCluster'] == 0)].sort_values('MonetaryValue', ascending=False) #Newbie - Your Newest Customers: First time buyers on your site. (2+0+x)
slip_cust = r_f_m[(r_f_m['RecencyCluster'] == 0) & (r_f_m['FrequencyCluster'] == 0)].sort_values('MonetaryValue', ascending=False) #Slip - Once Loyal, Now Gone: Great past customers who haven't bought in awhile. (0+0+x)
r_f_m['Status'] = 'Regular' 
r_f_m.loc[r_f_m['RFMScore'] == 6, 'Status'] = 'Core'
r_f_m.loc[r_f_m['FrequencyCluster'] == 2, 'Status'] = 'Loyal'
r_f_m.loc[r_f_m['MonetaryValueCluster'] == 2, 'Status'] = 'Whale'
r_f_m.loc[((r_f_m['RecencyCluster'] == 2) & (r_f_m['FrequencyCluster'] == 0)), 'Status'] = 'Newbie'
r_f_m.loc[((r_f_m['RecencyCluster'] == 0) & (r_f_m['FrequencyCluster'] == 0)), 'Status'] = 'Slip'
r_f_m.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

le = LabelEncoder()
r_f_m["Status"] = le.fit_transform(r_f_m["Status"])
X = r_f_m.drop(["Status"], axis=1)
y = r_f_m["Status"]
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer

dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
svm = SVC(random_state=42)
knn = KNeighborsClassifier()

metrics = {
    "accuracy": make_scorer(accuracy_score),
    "precision_micro": make_scorer(precision_score, average="micro"),
    "recall_micro": make_scorer(recall_score, average="micro"),
    "f1_micro": make_scorer(f1_score, average="micro"),
}

results = []
for clf in [dt, rf, svm, knn]:
    scores = cross_validate(clf, X, y, cv=10, scoring=metrics)
    results.append([scores[f"test_{metric}"].mean() for metric in metrics])

table = pd.DataFrame(results, index=["Decision Tree", "Random Forest", "SVM", "K-nearest neighbors"], columns=metrics.keys())
print(table)
```

```python
from sklearn.tree import export_graphviz
import graphviz

# Train a decision tree model
dt.fit(X, y)

# Export the decision tree to a DOT file
export_graphviz(dt, out_file="tree.dot", 
                feature_names=X.columns, 
                class_names=["Regular", "Loyal", "Newbie", "Slip", "Whale"], 
                filled=True)

# Render the DOT file using Graphviz
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# Train a Random Forest model
rf.fit(X, y)

# Extract the individual decision trees from the Random Forest
estimators = rf.estimators_

# Export each decision tree to a DOT file
for i, estimator in enumerate(estimators):
    export_graphviz(estimator, out_file=f"tree_{i}.dot", 
                    feature_names=X.columns, 
                    class_names=["Regular", "Loyal", "Newbie", "Slip", "Whale"], 
                    filled=True)

# Render the decision trees using Graphviz
trees = [graphviz.Source(open(f"tree_{i}.dot").read()) for i in range(len(estimators))]
graphviz.Source("\n".join(tree.source for tree in trees))
```

```python
import numpy as np

# Train a linear SVM classifier
svm = SVC(kernel='linear', random_state=42)
svm.fit(X, y)

# Get the support vectors and their indices
support_vectors = svm.support_vectors_
support_vectors_idx = svm.support_

# Plot the decision boundary and the support vectors
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap='viridis')
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], color='red')
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx = np.linspace(xlim[0], xlim[1], 10)
yy = np.linspace(ylim[0], ylim[1], 10)
XX, YY = np.meshgrid(xx, yy)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm.decision_function(xy).reshape(XX.shape)
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.show()
```

```python
import numpy as np

# Train a KNN classifier
knn.fit(X, y)

# Generate a grid of points to evaluate the classifier
h = .02
x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])

# Plot the Voronoi diagram and the data points
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10, 8))
plt.pcolormesh(xx, yy, Z, cmap='Pastel1')
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, edgecolor='k', cmap='viridis')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("KNN clustering with K=3")
plt.show()
```
