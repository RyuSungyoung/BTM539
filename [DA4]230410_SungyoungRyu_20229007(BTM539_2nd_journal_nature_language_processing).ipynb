```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

#PNLSW preprocessing function defined:
def preprocess_text(text):
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = ''.join(c for c in text if not c.isdigit())
    text = text.lower()
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    ps = PorterStemmer()
    text = ' '.join([ps.stem(word) for word in text.split()])
    return text

df['Preprocessed Abstract'] = df['Abstract'].apply(preprocess_text)
```

**Punctuation Removal:** Remove all punctuation marks from the text.<BR>
**Number Removal:** Remove all numeric characters from the text.<BR>
**Lowercase Conversion:** Convert all text to lowercase.<BR>
**Stopword Removal:** Remove all common stop words from the text.<BR>
**Stemming/Lemmatization:** Reduce all words to their root/stem form.<BR>

```python
import re

def word_count(text): #counting word~ word in abstract~ better than louis vuitton~ louis vuitton~
    words = re.findall(r'\w+', text)
    return len(words)

print(df['Abstract'].apply(word_count), df['Preprocessed Abstract'].apply(word_count))
```
