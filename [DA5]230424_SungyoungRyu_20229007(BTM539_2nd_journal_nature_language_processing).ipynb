```python
copy_index = []
for cor in corpus:
    if 'Copyright (c)' in cor:
        copy_index.append(corpus.index(cor))
    elif 'Copyright (C)' in cor:
        copy_index.append(corpus.index(cor))
    elif 'Research Summary' in cor:
        copy_index.append(corpus.index(cor))
    elif 'Research summary' in cor:
        copy_index.append(corpus.index(cor))
    elif 'Research Abstract' in cor:
        copy_index.append(corpus.index(cor))
    else:
        continue
```

```python
len(copy_index) #detecting copyright all reversed and research abstract or summary head by corpus iteration
```

```python
corpsn = [i.replace('Research Summary','').replace('Research Abstract','').replace('Research summary','') for i in corpus] #rip off research abstract or summary head
len(corpsn)
```

```python
import nltk
nltk.download('punkt')
from nltk.text import sent_tokenize
corpsnn = []
for i in corpsn:
    sentences = sent_tokenize(i)
    for j in sentences:
        if 'Copyright (' in j: #get rid of copyright all reversed tail
            sentences.remove(j)
    cleanedAbst = ' '.join(sentences)
    corpsnn.append(cleanedAbst)
```

```python
#LSA with Scikit-learn
P_names = ['p' + str(i) for i in range(len(corpus_PNLMW))] #add 'p' + all corpus_PNLMW index
pd.Series(P_names)
```

```python
def feed(wordlist):
    """Use tjhis for feeding preprocessed tokens"""
    return wordlist
#DTM = document-term matrix = word-document matrix(v x m: bag of words)
dtm_md = CountVectorizer(tokenizer = feed, min_df = 5, max_df = 0.5, ngram_range = (1,1),
                         token_pattern = None, lowercase = False)
dtfm = dtm_md.fit_transform(corpus_PNLMW)
print(type(dtfm), dtfm.shape)
```

```python
dtfmx = pd.DataFrame(dtfm.toarray(), index = P_names, columns = dtm_md.get_feature_names_out())
dtfmx
```

```python
#DTM to DTL(Document-Term List) of Gensim Library
fin_words = dtm_md.inverse_transform(dtfm)
print(len(fin_words[0]), fin_words[0])
```

```python
pp_dict = Dictionary(fin_words) #mapping into dictionary
print(pp_dict[4], pp_dict.token2id['ability']) #the 4th word in bag is "change", and "ability" is situated in 578th.
```

```python
pp_tdf = list(sorted(pp_dict.dfs.items())) #Tf-idf(term frequency and inverse document frequency)
tdf_pp = [(pp_dict[id], freq) for id, freq in sorted(pp_dict.dfs.items())]
print(tdf_pp) #Order by Term (a-b-c)
print(sorted(tdf_pp, key = lambda x: x[1], reverse = True), end=' ') #Order by Frequency
```

```python
wordfreq_doc = [pp_dict.doc2bow(text) for text in fin_words]
print(len(wordfreq_doc))
print(wordfreq_doc[0], end = " ")
print()
print([[(pp_dict[id], freq) for id, freq in cp] for cp in wordfreq_doc][0], end = ' ')
```

```python
#using a list to build an elbow chart to decide the optimal numbers of explained_variance_ratio_.sum()
elbowchart = []

for num_topics in range(5, 55, 5): #5~50 iteration, +5
    #fit LSA model
    pp_LSA_sklm = TruncatedSVD(n_components= num_topics, n_iter = 7, random_state = 42)
    pp_LSA_dtm = pp_LSA_sklm.fit_transform(dtfm)
    elbowchart.append(pp_LSA_sklm.explained_variance_ratio_.sum()) #add variance explained by LSA model to elbow chart list
    print(f'Variance explained by {num_topics} topics: {pp_LSA_sklm.explained_variance_ratio_.sum()}')

plt.plot(list(range(5, 55, 5)), elbowchart, marker='x')
plt.xlabel('Number of Topics')
plt.ylabel('Variance Explained')
plt.title('Variance Explained by Different Numbers of Topics')
plt.show()

#no dramatically difference by controlling over pp_LSA_sklm.explained_variance_ratio_.sum()
```
